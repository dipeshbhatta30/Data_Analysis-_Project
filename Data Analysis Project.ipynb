# Creating visualizations with Matplotlib

# Visualization of Yearly Average GDP

```python
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['Average_GDP'], label='Average GDP(Line)', color='Red', marker='o')
plt.scatter(df['year'], df['Average_GDP'], label='Average GDP(Scatter)', color='Green', marker='o')

plt.title('GDP Average From 2003 -2021')
plt.xlabel('Year')
plt.ylabel('Average GDP')
plt.legend()
plt.show()
```

# Visualization of Yearly Average Interest Rate

```python
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['Average_INTDSRUSM193N_Rate'], label='Average Interes tRate(Line)', color='Blue', marker='o')
plt.scatter(df['year'], df['Average_INTDSRUSM193N_Rate'], label='Average Interest Rate(Scatter)', color='Green', marker='o')


plt.title('Average Interest Rate From 2003 -2021')
plt.xlabel('Year')
plt.ylabel('Average Interest Rate')
plt.legend()
plt.show()
```

# Visualization of Yearly Median Sales Price

```python
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['Average_Median_Sales_Price'], label='Median Sales Price(Line)', color='Green', marker='o')
plt.scatter(df['year'], df['Average_Median_Sales_Price'], label='Median Sales Price(Scatter)', color='Red', marker='o')

plt.title('Average InterestRate From 2003-2021')
plt.xlabel('Year')
plt.ylabel('Median Sales Price')
plt.legend()
plt.show()
```
